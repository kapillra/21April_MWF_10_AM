# Legends, Grids, Axes, Labels

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
days = [*range(1, 8)]
```

```python
[np.random.randint(10, 99) * 1.2 / 2 for i in range(1, 8)]
```

```python
max_temp = [np.random.randint(10, 99) * 1.2 / 2 for i in range(1, 8)]
```

```python
min(max_temp)
```

```python
for i in range(1, 8):
    m = [np.random.randint(10, 99) * 1.2 / 2 for i in range(1, 8)]
    print(min(m))
```

```python
[min([np.random.randint(10, 99) * 1.2 / 2 for i in range(1, 8)]) for i in range(1,8)]
```

```python
min_temp = [min([np.random.randint(10, 99) * 1.2 / 2 for i in range(1, 8)]) for i in range(1,8)]
```

```python
min_temp
```

```python
zip(max_temp, min_temp)
```

```python
for i in zip(max_temp, min_temp):
    print(i)
```

```python
for i,j in zip(max_temp, min_temp):
    print(i,j)
```

```python
for i,j in zip(max_temp, min_temp):
    print(i+j/2)
```

```python
avg_temp = [(i+j)/2 for i,j in zip(max_temp, min_temp)]
```

```python
avg_temp
```

```python
max_temp, min_temp, avg_temp
```

```python
# plotting

plt.plot(days, max_temp, label='Max')
plt.plot(days, min_temp, label='Min')
plt.plot(days, avg_temp, label='Avg')
```

```python
# set axes labels and chart title
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_temp, label='Max')
plt.plot(days, min_temp, label='Min')
plt.plot(days, avg_temp, label='Avg')
```

```python
# Show legend

plt.plot(days, max_temp, label='Max')
plt.plot(days, min_temp, label='Min')
plt.plot(days, avg_temp, label='Avg')

plt.legend(loc='best')
```

```python
# Legend at different location with shadow enabled and fontsize set to large
plt.plot(days, max_temp, label="max")
plt.plot(days, min_temp, label="min")
plt.plot(days, avg_temp, label="average")

plt.legend(loc='upper right',shadow=True,fontsize='large')
```

```python
# shadow and grid

plt.plot(days, max_temp, label="max")
plt.plot(days, min_temp, label="min")
plt.plot(days, avg_temp, label="average")

plt.legend(loc='upper right', fontsize="large",shadow=True)
plt.grid()
```
