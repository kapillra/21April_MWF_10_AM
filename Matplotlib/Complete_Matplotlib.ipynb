```python
# importing libs
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [*range(1,8)]
```

```python
y = [np.random.randint(10, 99) for i in range(1, 8)]
```

```python
y
```

```python
len(x) == len(y)
```

```python
plt.xlabel("Days")
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(x, y, color='blue', linewidth=2, linestyle='dotted')
```

```python
# set cross points

plt.plot(x,y, 'b+--')
```

```python
plt.plot(x,y, 'b*--')
```

```python
plt.plot(x,y, 'b.--')
```

```python
# same effect as 'b+' format string

plt.plot(x,y, color='red', marker='*', linestyle='')
```

```python
# markersize

plt.plot(x,y,color='blue',marker='*',linestyle='',markersize=20)
```

```python
# alpha property to control line transperancy

plt.plot(x,y,'*',alpha=0.5) # alpha can be specified on a scale 0 to 1
```
