# Bar Chart

```python
# importing libs

import matplotlib.pyplot as plt
import numpy as np
```

```python
companies = ['INFOSYS', 'TCS', 'WIPRO', 'TECH MAHINDRA', 'HCL']
revenue = [55,99,60,77.9,55]
```

```python
xpos = np.arange(len(companies))
```

```python
xpos
```

```python
plt.bar(xpos, revenue, label='Revenue')
plt.xticks(xpos,companies)
plt.ylabel('Revenue [billion]')
plt.title('Indian IT Companies')
plt.legend()
```

```python
# Multiple Bars showing revenue and profit of major US tech companies

profit = [40,2,34,12,20]
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,companies)
plt.ylabel('Revenue [billion]')
plt.title('Indian IT Companies')
plt.legend()
```

```python
# showing bars horizontally

plt.barh(xpos-0.2,revenue, height=0.4, label="Revenue")
plt.barh(xpos+0.2,profit, height=0.4,label="Profit")

plt.yticks(xpos,companies)
plt.ylabel('Revenue [billion]')
plt.title('Indian IT Companies')
plt.legend()
```
